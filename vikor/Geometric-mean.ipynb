```python
import pandas as pd
import numpy as np
data = pd.read_csv('test.csv')
data.drop(columns=['Criteria'],inplace=True)
```

```python
data
```

```python
# calculating the float values
def division(x):
    try:
        if '/' in x:
            parts = x.split('/')
            return int(parts[0]) / int(parts[1])
        else:
            return int(x)
    except:
        return int(x)
def float_values(df):
    df_n = df.copy()
    try:
        df_n.drop(columns=['Pair-wise comp'],inplace=True)
    except:
        df_n = df_n
    for col in df_n.columns:
        df_n[col] = df_n[col].apply(lambda x:division(x))
    return df_n

def normalization(df):
    df_n = df.copy()
    for col in df.columns:
        sum_ = round(df_n[col].sum(),2)
        df_n[col] = df_n[col].apply(lambda x:round(x/sum_,2))
    return df_n
```

```python
cal_df = float_values(data)
norm_df = normalization(cal_df)
norm_df
```

```python
def geometic_calculation(df):
    df_n = df.copy()
    row_sums = df_n.sum(axis=1)
    result = np.power(row_sums, 1/16)
    return result
```

```python
def norm_weight(weight_list):
    weight_list_sum = sum(weight_list)
    result = []
    for i in weight_list:
        result.append(round((i/weight_list_sum),2))
    return result
```

```python
un_norm_weight = geometic_calculation(norm_df)
```
