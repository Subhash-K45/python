Outlier Handling

```python
Z_score = ['NacellePosition','AmbientTemperatue']
Percentile_99 = ['WindSpeed']
Percentile_1 = ['GeneratorRPM','HubTemperature','RotorRPM','ActivePower']
Percentile = ['BearingShaftTemperature','Blade2PitchAngle','Blade3PitchAngle','Blade1PitchAngle','GeneratorWinding1Temperature','GeneratorWinding2Temperature']
Quantile   = ['GearboxBearingTemperature','GearboxOilTemperature','MainBoxTemperature']
```

```python
for i in Z_score:
  mean = data[i].mean()
  std = data[i].std()
  z_score = (data[i]-mean)/std
  max_limit = mean + 3*std
  min_limit = mean - 3*std
  data[i] = np.where(data[i]<min_limit,min_limit,np.where(data[i]>max_limit,max_limit,data[i]))
```

```python
for i in Percentile_99:
  max_limit = data[i].quantile(0.99)
  data[i]=np.where(data[i]>max_limit,max_limit,data[i])
```

```python
for i in Percentile_1:
  min_limit = data[i].quantile(0.01)
  data[i]=np.where(data[i]<min_limit,min_limit,data[i])
```

```python
for i in Percentile:
  min_limit = data[i].quantile(0.01)
  max_limit = data[i].quantile(0.99)
  data[i] = np.where(data[i]<min_limit,min_limit,np.where(data[i]>max_limit,max_limit,data[i]))
```

```python
for i in Quantile:
  q1 = data[i].quantile(0.25)
  q3 = data[i].quantile(0.75)
  IQR = q3-q1
  min_limit = q1- 1.5*IQR
  max_limit = q3 + 1.5*IQR
  data[i] = np.where(data[i]>max_limit,max_limit,np.where(data[i]<min_limit,min_limit,data[i]))
```

```python
data.to_csv('/content/df_2019.csv',index=False)
```
