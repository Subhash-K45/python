```python
ls = [mapped_dm1, mapped_dm2, mapped_dm3]
result_data = []

for x in ls[0].columns:
    row_data = []
    for a, b, c in zip(mapped_dm1[x], mapped_dm2[x], mapped_dm3[x]):
        row_data.append([min(a[0], b[0], c[0]), round(sum([a[1], b[1], c[1]]) / 3, 2), max(a[2], b[2], c[2])])
    result_data.append(row_data)
```

```python
len(result_data)
```

```python
agg_data = pd.DataFrame(result_data,columns=['Liposomes',
'PLGA Nanoparticles',
'Silica Nanoparticles',
'Dendrimers',
'Nanoemulsions',
'Polymeric Micelles',
'Magnetic Nanoparticles'
],index = [mapped_dm1.columns]).T
agg_data
```

Weightage and beneficial Classification

```python
Weighted_Matrix = [(5,7,9),(5,7,9),(3,5,7),(1,3,5),(1,1,3),(7,7,9),(1,3,5),(1,1,3)]
beneficial = [0,1,1,0,1,0,1,1]
```

Finding the minimum and maximum according to beneficial criteria

```python
max_=-1
min_=9999
for (i,col) in enumerate(agg_data.columns):
  for val in agg_data[col]:
    if(beneficial[i]):
      max_ = max(max(val),max_)
    else:
      min_ = min(min(val),min_)
print(min_,max_)
```

```python
for i, col in enumerate(agg_data.columns):
    if beneficial[i]:
        agg_data[col] = agg_data[col].apply(lambda x: [round(val / max_, 2) for val in x])
    else:
        agg_data[col] = agg_data[col].apply(lambda x: [round( min_/val, 2) for val in x])
```

```python
agg_data
```

```python
result_df = pd.DataFrame()
for i, col in enumerate(agg_data.columns):
    result_df[col] = agg_data[col].apply(lambda x: [round(val * Weighted_Matrix[i][j], 2) for j, val in enumerate(x)])
result_df
```

```python
A_star = [[5.0, 7.0, 2.97],[2.8, 5.46, 9.0],[1.68, 3.9, 7.0],[1.0, 0.99, 0.7],[0.33, 0.7, 3.0],[7.0, 7.0, 2.97],[0.56, 2.34, 5.0],[0.56, 0.78, 3.0]]
A_minus=[[1.0, 0.98, 0.99],[0.55, 0.77, 2.97],[0.33, 2.05, 5.46],[0.2, 0.42, 0.55],[0.11, 0.11, 0.99],[2.31, 1.26, 0.99],[0.11, 0.57, 2.8],[0.11, 0.11, 0.99]]
```

```python
result_df.columns = [col[0] for col in result_df.columns]
```

```python
for i, x in enumerate(result_df.columns):
    print(result_df[x].apply(lambda val: [round(v * A_star[i][j], 2) for j, v in enumerate(val)]))
```

```python
distances_df_p = pd.DataFrame(index=result_df.index, columns=result_df.columns)
for i, x in enumerate(result_df.columns):
    distances_df_p[x] = result_df[x].apply(lambda val: round(np.sqrt((1/3) * np.sum((np.array(A_star[i]) - np.array(val))**2)),2))
distances_df_p
```

```python
distances_df_n = pd.DataFrame(index=result_df.index, columns=result_df.columns)
for i, x in enumerate(result_df.columns):
    distances_df_n[x] = result_df[x].apply(lambda val: round(np.sqrt((1/3) * np.sum((np.array(A_minus[i]) - np.array(val))**2)),2))
distances_df_n
```

```python
distances_df_n['row_sum'] = distances_df_n.sum(axis=1)
```

```python
distances_df_p['row_sum'] = distances_df_p.sum(axis=1)
```

```python
distances_df_p = distances_df_p.round(2)
```

```python
distances_df_n = distances_df_n.round(2)
```

```python
result_cc = pd.DataFrame({'d_star': distances_df_n['row_sum'], 'd_minus': distances_df_p['row_sum']})
result_cc
```

```python
result_cc['cci'] = result_cc['d_minus']/  result_cc['d_star']+result_cc['d_minus']
```
